# Outliers are present at both Lower and Upper Limit 
# Following the steps to remove outliers of rating columns

```python
# Finding Q1 & Q3

Q1,Q3 = df.rating.quantile([.25,.75])
Q1,Q3
```

```python
# Finding IQR
IQR = Q3-Q1
IQR
```

```python
# Finding Lower Limit (LL) and Upper Limit (UL)

LL = Q1-IQR*1.5
UL = Q3+IQR*1.5
LL,UL
```

```python
# Replacing the outliers with UL AND LL

df.rating = np.where(df.rating>UL,UL,df.rating)
df.rating = np.where(df.rating<LL,LL,df.rating)
```

```python
# Now checking the changes with boxplotting

sns.boxplot(df.rating,orient="h")
```
